### Impose an interest rate of 2.9% for each customer for any due amount.

```python
def intrest_rate(amount):
    amount_pay=amount*(2.9/100)
    return amount_pay+amount
```

```python
intrest_rate(8000)
```

```python
df3["due_amount"]=df3["Amount"].apply(intrest_rate)
```

```python
df3
```

###  Monthly profit for the bank.

```python
def profit(amount):
    amount_pay=amount*(2.9/100)
    return amount_pay
```

```python
df3["profit"]=df3["Amount"].apply(profit)
```
